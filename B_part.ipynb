```python
import pandas as pd
from sklearn import svm
from sklearn.model_selection import GridSearchCV
import os
from skimage.transform import resize
from skimage.io import imread
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pickle
from sklearn.decomposition import PCA
import plotly.express as px
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
dataDir = 'image'
cat = ['ice', 'Cars']
img_array = 'black-cone-7.jpg'
flat_data_arr = []
target_arr = []
for i in cat:
    path=os.path.join(dataDir,i)
    print(len(os.listdir(path)))
    for img in os.listdir(path):
        # print(img, path)
        # print(img)
        img_array = imread(os.path.join(path,img))
        img_resized = resize(img_array,(150,150,3))
        flat_data_arr.append(img_resized.flatten())
        target_arr.append(cat.index(i))

flat_data=np.array(flat_data_arr)
target=np.array(target_arr)
df=pd.DataFrame(flat_data)
df['Target']=target
df
        # img_resized=resize(img_array,)
    # for img in os.listdir(path):
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=77,stratify=y)
```

```python
pca = PCA(n_components=3)
pca.fit(x_train)
finals = pca.transform(x_train)

xshow=pd.DataFrame(data = finals , columns=['pc1' , 'pc2', 'pc3'])
xshow
```

```python
y_train.shape
```

```python
param_grid={'C':[0.1,1,10,100],'gamma':[0.0001,0.001,0.1,1],'kernel':['rbf','poly']}
svc=svm.SVC(probability=True)
model=GridSearchCV(svc,param_grid)

scores = cross_val_score(model,finals,y_train,cv=5)

print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
print("Number of CV Scores used in Average: ", len(scores))
```

```python
param_grid={'C':[0.1,1,10,100],'gamma':[0.0001,0.001,0.1,1],'kernel':['rbf','poly']}
svc=svm.SVC(probability=True)
model=GridSearchCV(svc,param_grid)
model.fit(finals,y_train)
model.best_params_
```
